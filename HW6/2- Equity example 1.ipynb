### Beta Calculation for Risk Assessment
Beta measures a stock's volatility in relation to the overall market. A beta greater than 1 indicates that the stock is more volatile than the market, while a beta less than 1 means it is less volatile.

- SPY (SPDR S&P 500 ETF Trust): SPY is one of the most widely recognized and traded ETFs in the world. It seeks to provide investment results that, before expenses, correspond generally to the price and yield performance of the S&P 500 Index. The S&P 500 Index is a broad-based measurement of changes in stock market conditions based on the average performance of 500 widely held common stocks.

- IVV (iShares Core S&P 500 ETF): Similar to SPY, IVV aims to track the investment results of an index composed of large-capitalization U.S. equities. It seeks to track the performance of the S&P 500 Index as well, offering another option for investors looking to invest in the top 500 companies in the U.S.

- VOO (Vanguard S&P 500 ETF): VOO is Vanguard’s ETF that tracks the S&P 500 Index. It aims to provide investors with the performance of the S&P 500, focusing on low costs and high efficiency, typical of Vanguard’s offerings.

- QQQ (Invesco QQQ Trust): QQQ is designed to track the investment results of the NASDAQ-100 Index, which includes 100 of the largest non-financial companies listed on the NASDAQ stock exchange. It is heavily weighted towards technology companies and is popular among investors looking for exposure to the tech sector.

- IWM (iShares Russell 2000 ETF): Unlike the others which focus on large-cap stocks, IWM aims to track the investment results of an index composed of small-capitalization U.S. equities. It seeks to track the performance of the Russell 2000 Index, making it a popular choice for investors looking for exposure to the small-cap segment of the U.S. equity market.

```python
market_etf = 'SPY'
betas = daily_returns.cov().loc[:, market_etf] / daily_returns.var()[market_etf]

plt.figure(figsize=(10, 6))
betas.drop(market_etf).plot(kind='bar')
plt.title('Beta Values of ETFs Relative to SPY')
plt.ylabel('Beta')
plt.xlabel('ETF')
plt.show()
```

### Portfolio Optimization
Portfolio optimization aims to maximize expected returns for a given level of risk, or equivalently, minimize risk for a given level of expected returns. Here, we'll use the Efficient Frontier concept.

The Efficient Frontier is a fundamental concept in modern portfolio theory introduced by Harry Markowitz in the 1950s. It represents a set of investment portfolios that offer the highest expected return for a given level of risk or the lowest risk for a given level of expected return. These portfolios are deemed "efficient" because they are designed to achieve maximum returns for a specified level of risk, making them optimal for investors.

Graphically, the Efficient Frontier is depicted as a curve on a plot with portfolio return on the y-axis and portfolio risk (typically measured by standard deviation or volatility) on the x-axis. Each point on the curve represents an efficient portfolio. The portfolios located below the curve are considered sub-optimal because they offer lower returns for a given level of risk, while those above the curve are theoretically not achievable due to the trade-off between risk and return.

Key characteristics of the Efficient Frontier include:

- Diversification: Portfolios on the Efficient Frontier typically involve diversification across a variety of asset classes to achieve the optimal balance between risk and return.

- Risk-Return Trade-off: It illustrates the principle of the risk-return trade-off, indicating that higher returns usually come with higher risk.

- Choice of Portfolio: Investors can choose a portfolio along the Efficient Frontier based on their risk tolerance. A more risk-averse investor might choose a portfolio towards the left of the curve (lower risk, lower return), while a risk-tolerant investor might select a portfolio towards the right (higher risk, higher return).

- Shifting Frontier: The Efficient Frontier can shift based on changes in market conditions, investment opportunities, and the addition of new assets into consideration, reflecting the dynamic nature of financial markets.

```python
from scipy.optimize import minimize

# Function to calculate portfolio return and volatility
def portfolio_performance(weights, mean_returns, cov_matrix):
    returns = np.dot(weights, mean_returns)
    std_dev = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
    return std_dev, returns

# Objective function to minimize (for a given target return)
def minimize_volatility(weights, mean_returns, cov_matrix, return_target):
    std_dev, returns = portfolio_performance(weights, mean_returns, cov_matrix)
    return std_dev

# Number of assets
num_assets = len(etfs)
mean_returns = daily_returns.mean()
cov_matrix = daily_returns.cov()

# Constraints and bounds
constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1},  # The sum of weights is 1
               {'type': 'eq', 'fun': lambda x: portfolio_performance(x, mean_returns, cov_matrix)[1] - return_target})
bounds = tuple((0, 1) for asset in range(num_assets))

# Optimize for a range of target returns
target_returns = np.linspace(mean_returns.min(), mean_returns.max(), 50)
target_volatilities = []

for return_target in target_returns:
    args = (mean_returns, cov_matrix, return_target)
    result = minimize(minimize_volatility, num_assets * [1./num_assets,], args=args,
                      method='SLSQP', bounds=bounds, constraints=constraints)
    target_volatilities.append(result['fun'])

plt.figure(figsize=(10, 6))
plt.scatter(target_volatilities, target_returns, c=target_returns / np.array(target_volatilities), marker='x')
plt.title('Efficient Frontier')
plt.xlabel('Volatility')
plt.ylabel('Return')
plt.colorbar(label='Sharpe Ratio')
plt.show()
```
